```python
# Making Changes to the data.
```

```python
# Adding the Total column.

df2['Total'] = df2['HP'] + df2['Attack'] + df2['Defense'] + df2['Sp. Atk'] + df2['Sp. Def'] + df2['Speed']

df2.drop(columns = ['Total'])

# Accessing a range of columns
df2['Total'] = df2.iloc[:, 4:10].sum(axis=1)
df2[['Total', 'HP', 'Defense']]

# Moving the position of columns
cols = list(df2.columns.values)
df2 = df2[cols[0:4] + [cols[-1]] + cols[4:12]]
df2
```

```python
df2.to_csv('Modified.csv', index=False)

df2.to_excel('Modified.xlsx', index=False)

# separators are ',' and '\t'
df2.to_csv('Modified.txt', index=False, sep='\t')
```

```python
# Filtering Data
```

```python
# Filtering data to obtain Type 1 == Grass and Type 2 == Poison
df2.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]

# Filtering data to obtain Type 1 == Grass or Type 2 == Poison 
df2.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]

# Filtering data to obtain Type 1 == Grass or Type 2 == Poison with HP > 70 
df2.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] < 70)]

# save to csv
df2.to_csv('filtered.csv', index=False)

# Assigning new index (resetting anyways)
new_df = df2.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] < 70)]
new_df.reset_index()

# Modifying it further( dropping previous index )
new_df.reset_index(drop=True, inplace=True)
new_df

# Filtering all data containing Mega.
df2.loc[df2['Name'].str.contains('Mega')]

# inverse of step above.
df2.loc[~df2['Name'].str.contains('Mega')]



# Using Regex
import re

df2.loc[df2['Type 1'].str.contains('Fire|Grass', regex = True)]

# also:
df2.loc[df2['Type 1'].str.contains('fire|grass',flags=re.I, regex = True)]

# Getting certain element in Name that start with 'pi'
df2.loc[df2['Name'].str.contains('^pi[a-z]*', flags=re.I, regex = True )]
```

```python
# Agregate Statistics.(Groupby)
df = pd.read_csv("pokemon_data.csv")
df.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
df.groupby(['Type 1']).count()
```

```python
# Working with large number of data.
```

```python
clean_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv("pokemon_data.csv", chunksize=200):
    results = df.groupby(['Type 1']).count()
    clean_df = pd.concat([clean_df, results])
    print(clean_df)
```
